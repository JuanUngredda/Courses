# Transformers Part II (BERTino)

This notebook is a **from-scratch encoder-only Transformer** (BERT-style) trained for **sentence classification** on the GLUE **SST-2** dataset (positive vs negative sentiment).

You’ll build the full pipeline:

- Load SST-2 from Hugging Face `datasets`
- Train a **simple character tokenizer** (with `[PAD]`, `[UNK]`, `[CLS]`)
- Create a PyTorch `Dataset` that returns:
  - `tokens` (padded to `T_max`)
  - `pad_mask` (1 for real tokens, 0 for `[PAD]`)
  - `label` (0/1)
- Implement the encoder stack:
  - token + positional embeddings
  - **non-causal** self-attention (with padding mask)
  - multi-head attention
  - MLP (feed-forward)
  - residuals + LayerNorm (pre-norm)
- Train with cross-entropy and evaluate loss/accuracy

> Compared to GPT (decoder-only), the encoder **can attend to all positions** (no causal mask). The only masking we do is to ignore padding tokens.

## 2) Train a simple BPE tokenizer

We build a small character tokenizer from the **training sentences only**.

Special tokens:
- `[PAD]` — used to pad sequences to a fixed length `T_max`
- `[UNK]` — used for any token not in the vocabulary
- `[CLS]` — prepended to every sentence; its final representation will be used for classification


```python
class CharTokenizer:
    def __init__(self, training_text: str):
        # Special tokens
        specials = ["[PAD]", "[UNK]", "[CLS]"]

        # Unique characters from training corpus
        chars = sorted(set(training_text))

        # Full vocabulary
        self.tokens = specials + chars

        # Mappings
        self.stoi = {tok: i for i, tok in enumerate(self.tokens)}
        self.itos = {i: tok for tok, i in self.stoi.items()}

        self.pad_id = self.stoi["[PAD]"]
        self.unk_id = self.stoi["[UNK]"]
        self.cls_id = self.stoi["[CLS]"]

    def get_vocab_size(self):
        return len(self.tokens)

    def encode(self, text: str):
        # Map each character to its id
        return [self.stoi.get(ch, self.unk_id) for ch in text]

    def decode(self, ids: list[int]):
        out = []
        for i in ids:
            tok = self.itos[i]
            if tok in ["[PAD]", "[CLS]"]:
                continue
            out.append(tok)
        return "".join(out)
```

### Build tokenizer from SST-2 training text

We lowercase the training sentences (matching your training setup) and join them into a big string.


```python
train_small = dataset["train"].shuffle(seed=0).select(range(10000))
val_small   = dataset["validation"]

training_sentences = train_small["sentence"]
training_sentences = [s.lower() for s in training_sentences]
training_sentences = "\n".join(training_sentences)

tokenizer = CharTokenizer(training_sentences)
print("Vocab size:", tokenizer.get_vocab_size())

# quick sanity check
sample = train_small[0]["sentence"].lower()
ids = tokenizer.encode(sample)
print("Sample:", sample)
print("Encoded length:", len(ids))
print("Decoded:", tokenizer.decode(ids[:50]))
```

## 3) Dataset wrapper for classification

We create a PyTorch `Dataset` that:
1. lowercases the sentence
2. tokenizes into ids
3. prepends `[CLS]`
4. truncates to `T_max`
5. pads with `[PAD]` to length `T_max`
6. returns a padding mask where **1 means real token** and **0 means padding**

The encoder uses this mask to avoid attending to `[PAD]` positions.


```python
class SST2Dataset(Dataset):

    def __init__(self, data, tokenizer, T_max):
        self.data = data
        self.tokenizer = tokenizer
        self.T_max = T_max

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        sample = self.data[index]
        ids, pad_mask = self.process_sample(sample["sentence"])
        return (torch.tensor(ids, dtype=torch.long),
                torch.tensor(pad_mask, dtype=torch.bool),
                torch.tensor(sample["label"], dtype=torch.long))

    def process_sample(self, text: str):
        # 1) lower case
        lowered_text = text.lower().strip()
        # tokenize
        ids = self.tokenizer.encode(lowered_text)
        # prepend CLS
        ids = [self.tokenizer.cls_id] + ids
        # truncate
        ids = ids[:self.T_max]
        # pad
        pad_len = self.T_max - len(ids)
        if pad_len > 0:
            ids = ids + [self.tokenizer.pad_id] * pad_len
        # 6) padding mask (1 for real tokens, 0 for PAD)
        pad_mask = [1 if t != self.tokenizer.pad_id else 0 for t in ids]
        return ids, pad_mask
```

```python
import numpy as np

lengths = [len(tokenizer.encode(s.lower().strip())) for s in train_small["sentence"]]

T_mean = float(np.mean(lengths))
T_p95  = float(np.percentile(lengths, 95))
T_max  = int(np.max(lengths))

print(T_mean, T_p95, T_max)
```

### Choose sequence length (`T_max`) and create DataLoaders

## 4) Encoder-only Transformer (BERTino)

Now we implement the model.

Differences vs GPT/decoder-only:
- **No causal mask** (tokens can attend to both left and right context)
- We apply a **padding mask** so attention ignores `[PAD]`
- For classification, we take the final hidden state at the `[CLS]` position and pass it through a linear head


```python
class EncoderEmbedding(nn.Module):
    def __init__(self, vocab_size: int, T_max: int, d_model: int, device: str):
        super().__init__()
        self.device = device
        self.T_max = T_max
        self.token_embedding = nn.Embedding(vocab_size, d_model, device=device)
        self.positional_embedding = nn.Embedding(T_max, d_model, device=device)

    def forward(self, input: Tensor) -> Tensor:
        # (batch_size, T_max)
        _, T = input.shape
        token_embedding = self.token_embedding(input)  # (batch_size, T_max, d_model)
        position_indeces = torch.arange(T, device=self.device)  # (T_max,)
        position_embedding = self.positional_embedding(position_indeces)  # (T_max, d_model)
        return token_embedding + position_embedding  # (batch_size, T_max, d_model)
```

### Self-attention head (padding-aware)

We compute attention scores `QKᵀ`, scale them by `sqrt(head_dim)`, and then **mask out key positions that are padding**.

Because this is an encoder, we do **not** apply a triangular causal mask.


```python
class Head(nn.Module):
    def __init__(self, d_model: int, head_dim: int, device: str, dtype=None) -> None:
        super().__init__()
        self.head_dim = head_dim
        self.Wq = nn.Linear(in_features=d_model, out_features=head_dim, bias=False, device=device, dtype=dtype)
        self.Wk = nn.Linear(in_features=d_model, out_features=head_dim, bias=False, device=device, dtype=dtype)
        self.Wv = nn.Linear(in_features=d_model, out_features=head_dim, bias=False, device=device, dtype=dtype)

    def forward(self, input: Tensor, pad_mask: Tensor) -> Tensor:
        # input: (batch_size, T_max, d_model)
        # pad_mask: (batch_size, T_max)
        Q = self.Wq(input)  # (batch_size, T_max, head_dim)
        K = self.Wk(input)  # (batch_size, T_max, head_dim)
        V = self.Wv(input)  # (batch_size, T_max, head_dim)
        scores = Q @ K.transpose(-2, -1)  # (batch_size, T_max, T_max)
        scores = scores / self.head_dim ** 0.5
        key_mask = pad_mask.unsqueeze(1)
        masked_attention_scores = scores.masked_fill(key_mask == 0, float("-inf"))
        softmax_weights = torch.softmax(masked_attention_scores, axis=-1)
        return softmax_weights @ V  # (batch_size, T_max, head_dim)
```

### Multi-head attention

We run multiple heads in parallel, concatenate their outputs, and project back to `d_model`.


```python
class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, n_heads, device, dtype) -> None:
        super().__init__()
        assert d_model % n_heads == 0, "d_model must be divisible by n_heads"
        heads_list = []
        for _ in range(n_heads):
            head = Head(d_model=d_model, head_dim=d_model // n_heads, device=device, dtype=dtype)
            heads_list.append(head)
        self.multihead_attention = nn.ModuleList(heads_list)
        self.Wout = nn.Linear(in_features=d_model, out_features=d_model)

    def forward(self, input: Tensor, pad_mask: Tensor) -> Tensor:
        # input: (batch_size, T_max, d_model)
        # pad_mask: (batch_size, T_max)
        head_outputs = [head(input, pad_mask) for head in self.multihead_attention]
        concat = torch.cat(head_outputs, dim=-1)  # (batch_size, T, d_model)
        return self.Wout(concat)  # (batch_size, T, d_model)
```

### Feed-forward network (MLP)

This is the standard Transformer MLP: expand to `4*d_model`, GELU, project back, dropout.


```python
class MLP(nn.Module):
    def __init__(self, d_model, device):
        super().__init__()
        self.fc1 = nn.Linear(d_model, 4 * d_model, device=device)
        self.act = nn.GELU()
        self.fc2 = nn.Linear(4 * d_model, d_model, device=device)
        self.dropout = nn.Dropout(0.1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.act(x)
        x = self.fc2(x)
        x = self.dropout(x)
        return x
```

### Encoder block (pre-norm)

Each block applies:

- `x = x + Dropout(MHA(LN(x)))`
- `x = x + Dropout(MLP(LN(x)))`

This matches the stable **pre-norm** pattern (LayerNorm before each sub-layer).


```python
class EncoderTransformerBlock(nn.Module):
    def __init__(self, d_model: int, n_heads: int, dropout_rate: float, device=None, dtype=None) -> None:
        super().__init__()
        self.attention = MultiHeadAttention(d_model=d_model, n_heads=n_heads, device=device, dtype=dtype)
        self.ln1 = nn.LayerNorm(d_model, device=device)
        self.ln2 = nn.LayerNorm(d_model, device=device)
        self.dropout = nn.Dropout(dropout_rate)
        self.ffn = MLP(d_model=d_model, device=device)

    def forward(self, x: Tensor, pad_mask: Tensor) -> Tensor:
        # (batch_size, T, d_model)
        x = x + self.dropout(self.attention(self.ln1(x), pad_mask))  # (batch_size, T, d_model)
        x = x + self.dropout(self.ffn(self.ln2(x)))  # (batch_size, T, d_model)
        return x  # (batch_size, T, d_model)
```

### Full encoder model + classification head

We stack `n_layers` encoder blocks, apply a final LayerNorm, then:

- take the hidden state at position `0` (the `[CLS]` token)
- map it to class logits with a linear layer


```python
class BERTino(nn.Module):
    def __init__(self, vocab_size: int,
                 T_max: int,
                 n_layers: int,
                 d_model: int,
                 n_heads: int,
                 dropout_rate: float,
                 num_classes: int,
                 device=None, dtype=None) -> None:
        super().__init__()
        self.embedding = EncoderEmbedding(vocab_size=vocab_size, T_max=T_max, d_model=d_model, device=device)
        self.encoder_stack = nn.ModuleList([EncoderTransformerBlock(d_model=d_model, n_heads=n_heads,
                                                                    dropout_rate=dropout_rate, device=device,
                                                                    dtype=dtype) for _ in range(n_layers)])

        self.ln = nn.LayerNorm(d_model, device=device, dtype=dtype)
        self.head = nn.Linear(d_model, num_classes, device=device, dtype=dtype)

    def forward(self, tokens: Tensor, pad_mask: Tensor) -> Tensor:
        x = self.embedding(tokens)  # (batch_size, T, d_model)
        for encoder in self.encoder_stack:
            x = encoder(x, pad_mask)  # (batch_size, T, d_model)
        x = self.ln(x)  # (batch_size, T, d_model)
        cls = x[:, 0, :]  # (batch_size, d_model)
        logits = self.head(cls)
        return logits # (batch_size, num_classes)
```

## 5) Instantiate model + optimizer

We now create the encoder with your configuration:

- `n_layers=3`
- `d_model=256`
- `n_heads=4`
- dropout `0.1`
- 2 classes for SST-2

Optimizer: `AdamW(lr=3e-4, weight_decay=0.01)` (standard Transformer choice).


```python
model = BERTino(vocab_size=tokenizer.get_vocab_size(),
                T_max=T_max,
                n_layers=3,
                d_model=256,
                dropout_rate=0.1,
                num_classes=2,
                n_heads=4,
                device=device, dtype=torch.float32).to(device)

print(model)
print("num params:", sum(p.numel() for p in model.parameters()))

optimizer = AdamW(model.parameters(), lr=3e-4, weight_decay=0.01)
```

## 6) Evaluation helper (loss + accuracy)

We periodically evaluate on a limited number of batches to track progress faster.

Notes:
- `model.eval()` disables dropout
- `torch.no_grad()` avoids building a computation graph
- accuracy is computed by `argmax` over logits


```python
@torch.no_grad()
def eval_loss_acc(loader, max_batches=5):
    model.eval()
    losses, accs = [], []
    for i, (tokens, masks, labels) in enumerate(loader):
        if i >= max_batches: break
        tokens, masks, labels = tokens.to(device), masks.to(device), labels.to(device)
        logits = model(tokens, masks)
        loss = F.cross_entropy(logits, labels)
        preds = logits.argmax(dim=-1)
        acc = (preds == labels).float().mean()
        losses.append(loss.item())
        accs.append(acc.item())
    model.train()
    return sum(losses)/len(losses), sum(accs)/len(accs)
```

## 7) Training loop

For each batch:
1. forward pass → logits
2. cross-entropy loss
3. backward pass
4. optimizer step

After each epoch, we print train/val loss (and you can also print accuracy if you want).


```python
num_epochs = 10
for epoch in range(num_epochs):
    i = 0
    for tokens, masks, labels in train_loader:
        tokens, masks, labels = tokens.to(device), masks.to(device), labels.to(device)
        logits = model(tokens, masks)
        loss = F.cross_entropy(logits, labels)
        optimizer.zero_grad(set_to_none=True)
        loss.backward()
        optimizer.step()
        print(i ," steps")
        i+=1
    tr_loss, tr_acc = eval_loss_acc(train_loader, max_batches=5)
    va_loss, va_acc = eval_loss_acc(val_loader, max_batches=5)
    print(f"epoch {epoch} | train loss {tr_loss:.4f} acc {tr_acc:.4f} | val loss {va_loss:.4f} acc {va_acc:.4f}")
```

## 8) Quick prediction (inference)

Let’s run the trained model on a few validation examples.

We:
- tokenize + pad the sentence
- run the model to get logits
- convert to probabilities with softmax (optional)
- pick the predicted class with `argmax`


```python
id2label = {0: "negative", 1: "positive"}

model.eval()
for i in range(5):
    sent = val_small[i]["sentence"]
    true = val_small[i]["label"]
    ids, pad_mask = val_ds.process_sample(sent)
    tokens = torch.tensor([ids], dtype=torch.long, device=device)
    masks = torch.tensor([pad_mask], dtype=torch.bool, device=device)

    with torch.no_grad():
        logits = model(tokens, masks)
        probs = torch.softmax(logits, dim=-1)[0]
        pred = int(probs.argmax().item())

    print("sentence:", sent)
    print("true:", id2label[true], "| pred:", id2label[pred], "| probs:", probs.detach().cpu().tolist())
    print("-" * 80)
```
